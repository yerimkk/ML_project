### EXPORT

```python
## 일단 수출중량이 0인 경우 제외한 export 데이터만 가져오기 

df_export = domestic_import_export_status[['연도', '분류', '광종', '수출중량(톤)', '수출금액(천불)']]
df_export = df_export[df_export['수출중량(톤)']>0]
```

```python
df_export['광종'].unique()
```

```python
df_export['분류'].unique()
```

```python
# 니켈, 리튬, 망간, 코발트 : 금속광
# 4개 광물 모두 금속광에 속하므로 금속광에 한정해서 분석

df_export = df_export[df_export['분류']=='금속광']
df_export_pivot = df_export.pivot_table(index=['연도', '광종'], values=['수출중량(톤)', '수출금액(천불)'], aggfunc='sum')
```

```python
# 망간이 없는 것 같아서 보니까
# 2022년에만 있다 (수출중량 0은 제외했음)
domestic_import_export_status[domestic_import_export_status['광종']=='망간광']
```

```python
# 코발트도 없다
# 수출중량 0이라서 제외된 것 (수출은 안한 것 같음)
domestic_import_export_status[domestic_import_export_status['광종']=='코발트']
```

```python
# 수출금액 
export_sum21_p = df_export[df_export['연도']==2021]['수출금액(천불)'].sum()
export_sum22_p = df_export[df_export['연도']==2022]['수출금액(천불)'].sum()

# 수출중량
export_sum21_q = df_export[df_export['연도']==2021]['수출중량(톤)'].sum()
export_sum22_q = df_export[df_export['연도']==2022]['수출중량(톤)'].sum()
```

```python
print('2021년 금속광 총 수출액', export_sum21_p)
print('2022년 금속광 총 수출액', export_sum22_p)
print('2021년 금속광 총 수출중량', export_sum21_q)
print('2022년 금속광 총 수출중량', export_sum22_q)
```

```python
df_export_pivot.loc[2021, '연도별 총합계(천불)'] = export_sum21_p
df_export_pivot.loc[2022, '연도별 총합계(천불)'] = export_sum22_p
df_export_pivot.loc[2021, '연도별 총합계(톤)'] = export_sum21_q
df_export_pivot.loc[2022, '연도별 총합계(톤)'] = export_sum22_q
```

```python
df_export_pivot['수출액비율(%)'] = df_export_pivot['수출금액(천불)']/df_export_pivot['연도별 총합계(천불)']*100
df_export_pivot['수출중량비율(%)'] = df_export_pivot['수출중량(톤)']/df_export_pivot['연도별 총합계(톤)']*100
```

```python
df_export_pivot['수출중량비율(%)']=round(df_export_pivot['수출중량비율(%)'], 4)
```

```python
# df_export_pivot.sort_values(by=['연도','수출액비율(%)'],ascending=[True,False], inplace=True)
# 일단은 수출 중량비율로만 확인
df_export_pivot.sort_values(by=['연도','수출중량비율(%)'],ascending=[True,False], inplace=True)
```
